```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline      
```

```python
x = np.arange(0,20)
y = x**2 +5
x,y
```

```python
#Creates linear plots with given y points with respect to x points
plt.plot(x,y)

#Adds grid to the system
plt.grid(True)

#Adds title to the visualization
plt.title("Title of this graph")
#Adds x-axis to the visualization
plt.xlabel("X Values")
#Adds y-axis to the visualization
plt.ylabel("Y Values")

#Define limits to show in graph
plt.xlim(0,25)
plt.ylim(0,400)
# Saving the figure
plt.savefig("output.png")
plt.show()
plt

```
